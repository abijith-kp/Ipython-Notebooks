Game of Life
=======

Simple script to run [Conway's Game of Life](https://en.wikipedia.org/wiki/Conway's_Game_of_Life) using [Ipythonblock](http://ipythonblocks.org/) python module.



```python
from ipythonblocks import BlockGrid
from itertools import product

def is_alive(block):
    return block.green == 255

def check_limit(x, y):
    if x >=0 and x < N and y >= 0 and y < N:
        return True
    else:
        return False
    
def get_neighbours(x, y, _grid):
    from itertools import product
    X = [x+1, x, x-1]
    Y = [y+1, y, y-1]
    xy = [(i, j)    for i, j in product(X, Y)    if check_limit(i, j)]
    xy.remove((x, y))
    return [_grid[i, j]    for i, j in xy]

def n_alive_dead(neighbours):
    alive = 0
    dead = 0
    for n in neighbours:
        if is_alive(n):
            alive += 1
        else:
            dead += 1
    return alive, dead

def get_grid(grid):
    _grid = grid.copy()
    total_alive = 0
    for block in _grid:
        neighbours = get_neighbours(block.row, block.col, _grid)
        alive, dead = n_alive_dead(neighbours)
        if is_alive(block):
            total_alive += 1
            if alive < 2 or alive >= 4:
                grid[block.row, block.col].green = 0
                total_alive -= 1
                #print('dead', (block.row, block.col))
        else:
            if alive == 3:
                grid[block.row, block.col].green = 255
                total_alive += 1
                #print('born', (block.row, block.col))
    if total_alive == 0:
        for i, j in seed:
            grid[i, j].green = 255
    return grid
    
def draw(grid):
    while True:
        grid = get_grid(grid)
        grid.flash(1)


N = 20
grid = BlockGrid(N, N)

init_seed = [(2, 3), (2, 4), (3, 5), (4, 5), (5, 5), (6, 5), (7, 5), (6, 2), (3, 6), (8, 5), (9, 5), (4, 3), (5, 3), 
             (6, 3), (7, 3), (8, 3), (9, 3), (10, 3), (10, 4), (1, 1), (1, 4), (11, 5), (10, 5), (4, 4), (5, 4), 
             (6, 4), (7, 4), (8, 4)]
seed = [(i+3, j+5)    for i, j in init_seed]

for i, j in seed:
    grid[i, j].green = 255
    
        
grid.block_size = 10
draw(grid)
```
